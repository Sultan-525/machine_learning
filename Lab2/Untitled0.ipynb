```python
import pandas as pd
import os # Import os module if you need to check file existence programmatically, though not strictly used in this try-except

# Define the filename
filename = 'store_sales.csv'

try:
    # 1. Load the dataset
    df = pd.read_csv(filename)

    # 2. Display shape
    print("Dataset Shape:", df.shape)

    # 3. Preview first few rows
    print("\nFirst 5 rows:")
    display(df.head())

    # 4. Check column names and data types
    print("\nColumns and Data Types:")
    print(df.dtypes)

except FileNotFoundError:
    print(f"Error: The file '{filename}' was not found.")
    print("Please ensure the file is uploaded to your Colab environment (e.g., using the 'Files' icon on the left sidebar)")
    print("or that the correct path to the file is provided if it's in a different location.")
    df = None # Set df to None to indicate loading failed
except Exception as e:
    print(f"An unexpected error occurred: {e}")

if df is None:
    print("\nDataFrame could not be loaded due to the above error. Please resolve the file issue and rerun the cell.")
```
