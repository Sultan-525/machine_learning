```python
from graphviz import Digraph


dot = Digraph('Methodology')
dot.attr(rankdir='TB')


dot.node('A', 'Start: Select Dataset\n(store_sales.csv)', shape='oval')
dot.node('B', 'Data Loading & Inspection\n(Pandas)', shape='box')
dot.node('C', 'Preprocessing\n(Cleaning/Encoding)', shape='box')
dot.node('D', 'Model Training\n(Regression)', shape='box')
dot.node('E', 'Evaluation\n(Results)', shape='oval')


dot.edge('A', 'B')
dot.edge('B', 'C')
dot.edge('C', 'D')
dot.edge('D', 'E')


dot.render('methodology', format='png')
dot
```
